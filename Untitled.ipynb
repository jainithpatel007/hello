```python
fig, ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.04))


ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(3, -6),
           arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"))
```

```python
import numpy as np
import matplotlib.pyplot as plt
#fig = plt.figure()

#syntax for 3D projection 

ax = plt.axes(projection = '3d')

#defining all 3 axes

.
z= np.linspace(0,10000,100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)

#3d ploting

ax.plot3D(x, y, z ,'green')
ax.set_title('3D')
plt.savefig("class.pdf",format = 'pdf')
```
